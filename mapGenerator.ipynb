```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import geoplot as gplt
import warnings
```

```python
geoHomeless = gpd.read_file('data/allData/completedHomelessData.zip')
homelessIndex = geoHomeless[['NAME', 'iH', 'geometry']]
homelessIndex
```

```python

homelessIndex = homelessIndex.explode()
homelessIndex['area'] = homelessIndex['geometry'].area
homelessIndex = homelessIndex.loc[homelessIndex.groupby('NAME')['area'].idxmax()].reset_index(drop=True)

#homelessIndex = homelessIndex.groupby('NAME').first().reset_index()

gplt.choropleth(homelessIndex, hue=homelessIndex['iH'], cmap='OrRd', figsize=(20, 20), legend=True)
homelessIndex
```

```python


homelessIndex['coords'] = homelessIndex['geometry'].apply(lambda x: x.centroid.coords[:])
homelessIndex['coords'] = [coords[0] for coords in homelessIndex['coords']]
midwayCoords = homelessIndex.at[23, 'coords']
homelessIndex.at[41, 'coords'] = (-118.015, 33.76229280131335)
homelessIndex.at[28, 'coords'] = (-117.85, 33.803795846706294)
homelessIndex.at[1, 'coords'] = (-117.9, 33.838974302280455)
map = gplt.choropleth(homelessIndex, hue=homelessIndex['iH'], cmap='OrRd', figsize=(20, 20), legend=True)
for idx, row in homelessIndex.iterrows():
    plt.annotate(text=row['NAME'], xy=row['coords'],
                 horizontalalignment='center', color='green', fontsize =12)
fig = map.figure
cb_ax = fig.axes[1] 

for label in cb_ax.get_yticklabels():
    label.set_color("white")  # Change 'blue' to your desired colorplt.legend(prop={'color':'blue'})
    label.set_fontsize(20)
plt.savefig('map.png', transparent=True)



```
